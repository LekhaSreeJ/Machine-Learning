```python
# importing the dataset from the local folder

train_norm = "D:\\AIML SEM 4\\python\\project\\chest_xray\\train\\NORMAL"
train_pneumo = "D:\\AIML SEM 4\\python\\project\\chest_xray\\train\\PNEUMONIA"
val_norm = "D:\\AIML SEM 4\\python\\project\\chest_xray\\val\\NORMAL"
val_pneumo ="D:\\AIML SEM 4\\python\\project\\chest_xray\\val\\PNEUMONIA"
test_norm = "D:\\AIML SEM 4\\python\\project\\chest_xray\\test\\NORMAL"
test_pneumo = "D:\\AIML SEM 4\\python\\project\\chest_xray\\test\\PNEUMONIA"
```

```python
# defining a function to convert image data into list of arrays of pixel values

def to_array(folder):
    train =[]
    for i in os.listdir(folder):
        gray()
        image = array(Image.open(folder+'\\'+i).resize((80,80)).convert('L'))
        x = image.reshape(1,-1)
        train.append(x)
    return train
```

**converting all the raw images to array of pixel values**

```python
n_train = to_array(train_norm)
```

```python
p_train = to_array(train_pneumo)
```

```python
n_val = to_array(val_norm)
```

```python
p_val = to_array(val_pneumo)
```

```python
n_test = to_array(test_norm)
```

```python
p_test = to_array(test_pneumo)
```

```python
#defining function to create dataframe of normal images

def dataframe_normal(array_normal):
    df = pd.DataFrame(data = (list(x) for x in array_normal))
    df = pd.DataFrame(df[0].to_list(),columns=range(0,6400))
    df['target']=0
    return df   
```

```python
#defining function to create dataframe of pneumonia images

def dataframe_pneumonia(array_pneumonia):
    df = pd.DataFrame(data = (list(x) for x in array_pneumonia))
    df = pd.DataFrame(df[0].to_list(),columns=range(0,6400))
    df['target']=1
    return df   
```

**converting to dataframe and concat them to create train and test data**

```python
data = [dataframe_normal(n_train), dataframe_pneumonia(p_train)]
```

```python
train = pd.concat(data)
```

```python
data1 = [dataframe_normal(n_val), dataframe_pneumonia(p_val)]
```

```python
train1 = pd.concat(data1)
```

```python
train_data = pd.concat([train,train1])
train_data
```

```python
data2 = [dataframe_normal(n_test), dataframe_pneumonia(p_test)]
```

```python
test_data = pd.concat(data2)
test_data
```

```python
# shuffling the dataset randomly 

from sklearn.utils import shuffle
train_data = shuffle(train_data,random_state = 56)
test_data = shuffle(test_data,random_state = 43)
```
